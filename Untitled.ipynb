## Final expression for estimation of the surface dipole
### $d = 1.59((-0.64222\chi + C_1) \times(-0.08056A + C_2)) + 0.06$
#### Build a function to do that in future ...

```python
def elec_surf_dipole(chi, A):
    return dipole_slope*((eNeg_slope*chi + eNeg_const)*(area_slope*A + area_const))
```

## Expression of interface dipole in terms of bulk $\chi$ and interface area values
### $D = -8.69101978691(\Delta d) + 0.3278$
### $\Delta d = k(\chi_1 -\chi_2)A$

```python
def interface_dipole(chi1,chi2,area):
    delta = -abs(-0.0305615528532*(abs(chi1 - chi2))*area)
    return -8.69101978691*delta + 0.3278
```
